```python
data['date/time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data['Base']
```

```python
def get_dom(dt):
    return dt.day

data['day'] = data['date/time'].map(get_dom)
```

```python
def get_month(dt):
    return dt.month

data['month'] = data['date/time'].map(get_month)
```

```python
def get_year(dt):
    return dt.year
data['year'] = data['date/time'].map(get_year)
```

```python
def get_hour(dt):
    return dt.hour
data['hour'] = data['date/time'].map(get_hour)
```

```python
def get_time(dt):
    return dt.time()

data['time_stamp'] = data['date/time'].map(get_time)
```

```python
def get_weekday(dt):
    return dt.weekday()
data['weekday'] = data['date/time'].map(get_weekday)
```
